```python
import pandas as pd
import  numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("UberDataset.csv")
```

```python
dataset
```

```python
dataset.info()
```

**Data Preprocesing**

```python
dataset['PURPOSE'].fillna("NOT" , inplace = True)
```

```python
dataset.head()
```

```python
dataset['START_DATE'] = pd.to_datetime(dataset['START_DATE'], errors = 'coerce' )


dataset['END_DATE'] = pd.to_datetime(dataset['END_DATE'], errors = 'coerce' )
```

```python
dataset.info()
```

```python
from datetime import datetime

dataset['DATE'] = pd.DatetimeIndex(dataset['START_DATE']).date

dataset['TIME'] = pd.DatetimeIndex(dataset['START_DATE']).hour
```

```python
dataset.head()
```

```python
dataset['DAY-NIGHT'] = pd.cut(x=dataset['TIME'],bins =[0,10,15,19,24],labels= ['Morning','Afternoon','Evening','Night'])
```

```python
dataset.head()
```

```python
dataset.dropna(inplace = True)
```

```python
dataset.shape
```

**Data Visualization**

```python
plt.figure(figsize=(20,5))

plt.subplot(1,2,1)

plt.yticks(rotation = 180)
sns.countplot(dataset['CATEGORY'])

plt.subplot(1,2,2)
sns.countplot(dataset['PURPOSE'])
```

```python
sns.countplot(dataset['DAY-NIGHT'])
```

```python
dataset.head()
```

```python
dataset['MONTH'] = pd.DatetimeIndex(dataset['START_DATE']).month

month_label = {1.0:'Jan',2.0:'Feb', 3.0:'March',4.0:'April',5.0:'May',
               6.0:'June',7.0:'July',8.0:'Aug',9.0:'Sept',10.0:'Oct',
               11.0:'Nov',12.0:'Dec' }

dataset["MONTH"] = dataset.MONTH.map(month_label)

mon =dataset.MONTH.value_counts(sort=False)
```

```python
dataset.head()
```

```python
df = pd.DataFrame({"Month":mon.values,
                  "Value Count":dataset.groupby("MONTH", sort = False)['MILES'].max()
                  })

p = sns.lineplot(data=df)
p.set(xlabel='MONTH' , ylabel='VALUE COUNT')
```

```python
dataset.head()
```

```python
dataset['DAY'] = dataset.START_DATE.dt.weekday


DAY_label = {0.0:'Monday',1.0:'Tuesday',2.0:'Wednesday',3.0:'Thursday',4.0:'Friday',
               5.0:'Saturday',6.0:'Sunday' }

dataset["DAY"] = dataset.DAY.map(DAY_label)
```

```python
dataset.head()
```

```python
day_label =dataset.DAY.value_counts()

sns.barplot(x=day_label.index,y= day_label)
plt.xlabel("DAY")
plt.ylabel(" COUNT")
```
